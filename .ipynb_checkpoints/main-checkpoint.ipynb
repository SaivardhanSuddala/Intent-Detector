```python
#importing all useful libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Understanding features of my dataset
dataset=pd.read_csv("train.csv")
print(dataset.columns)
labels_using=[selected_intents = [
    "card_arrival",
    "card_not_working",
    "lost_or_stolen_card",
    "pin_blocked",
    "card_payment_not_recognised",
    "transfer_not_received_by_recipient",
    "pending_top_up",
    "balance_not_updated_after_cheque_or_cash_deposit",
    "extra_charge_on_statement",
    "cancel_transfer"
]
dataset=dataset[dataset['category'].isin(labels_using)]
print(len(dataset))
print(dataset['category'].unique())
x=dataset['text']
y=dataset['category']
print(x.head(),y.head())
```

```python
#Shuffling datasets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2,stratify=y)
```

```python
#Model
from sklearn.linear_model import SGDClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
model=SGDClassifier(random_state=42)
vect=TfidfVectorizer(ngram_range=(1,2),max_features=5000)
x_train_vec=vect.fit_transform(x_train)
x_test_vec=vect.transform(x_test)
model.fit(x_train_vec,y_train)
result=model.predict(x_test_vec)
accuracy=accuracy_score(result,y_test)
print(accuracy)

```
